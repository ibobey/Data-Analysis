<center> <h3 style="color:magenta"> Altering Data </h3> </center>

```python
dataset['TotalPower'] = dataset.HP + dataset.Defense + dataset.Speed
```

```python
dataset.sort_values(['TotalPower'], ascending=0)
```

<center> <h3 style="color:magenta"> Filtering Data </h3> </center>

```python
dataset.loc[(dataset['Type 1'] == 'Normal') & (dataset.TotalPower > 120)]
```

<center> <h3 style="color:magenta"> Conditional Changes </h3> </center>

```python
dataset.loc[dataset['TotalPower'] > 300, ['Generation', 'Legendary']] = ['Test 1', 'Test 2']
```

```python
dataset.sort_values(['TotalPower'], ascending=0)
```
