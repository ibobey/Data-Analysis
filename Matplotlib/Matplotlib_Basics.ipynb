<center> <h2 style="color:blue"> Multiple Graphs </h2> </center>

```python
x_data = [1, 2, 3, 4, 5]
y_data_1 = [1, 2, 3, 4, 5]
y_data_2 = [1, 4, 9, 16, 25]
y_data_3 = [1.5, 2.17, 7, 5, 14]
```

```python
plt.plot(x_data, y_data_1, "r")
plt.plot(x_data, y_data_2, "b")
plt.scatter(x_data, y_data_3, color="g")
plt.show()
```

<h2 style="color:dodgerblue">Line Plot </h2> 

```python
x = np.arange(0,1,0.01)
y = np.sqrt(x)
```

```python
plt.plot(x, y)
plt.plot(x,x)
plt.plot(x, x**2)
plt.show()
```

<h2 style="color:dodgerblue">Line Plot </h2> 

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 35]
```

```python
plt.bar(categories, values)
plt.show()
```

<h2 style="color:dodgerblue">Scatter Plot </h2> 

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [0.7, 3, 4.4, 3, 5.2]
```

```python
plt.scatter(x, y)
plt.plot(x, x)
plt.show()
```

<h2 style="color:dodgerblue">Pie Plot </h2> 

```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Plot Example')
plt.axis('equal')
plt.show()
```

<h2 style="color:dodgerblue">Area Plot </h2> 

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]
```

```python
plt.fill_between(x, y1, color="red", alpha=0.4)
plt.fill_between(x, y2, color="dodgerblue", alpha=0.4)
plt.show()
```
